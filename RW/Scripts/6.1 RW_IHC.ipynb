```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 999)
```

```python
# load data frame
rw = pd.read_csv("..//Datasets/NEW/RW_with_airports.csv")
```

```python
rw
```

```python
# take only IHC service
ihc = rw.loc[rw.Owner == 'IHC HEALTH SERVICES INCSALT LAKE CITY, UT, US(Corporation)'].reset_index(drop=True)
```

```python
ihc['day_name'] = pd.to_datetime(ihc.date, format='%Y-%m-%d').dt.day_name()
```

```python
ihc['weekends'] = np.where((ihc.day_name == 'Saturday') | (ihc.day_name == 'Sunday'), 1, 0)
```

```python
ihc.head()
```

```python
ihc.to_csv('..//Datasets/NEW/RW_ihc.csv', index=False)
```
